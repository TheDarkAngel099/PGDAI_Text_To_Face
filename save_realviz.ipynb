```python
import torch
from diffusers import DiffusionPipeline
import matplotlib.pyplot as plt
```

```python
checkpoint_path = "/content/checkpoint"
```

```python
# Load RealVisXL pipeline from Hugging Face model hub
pipe = DiffusionPipeline.from_pretrained(
    "SG161222/RealVisXL_V4.0",
    torch_dtype=torch.float32,
    use_safetensors=True,
    verbose=False,
)
pipe.load_lora_weights(checkpoint_path)
```

```python
########## PKL ############
import pickle

# Save the loaded DiffusionPipeline model as a pickle file
with open('RealVisXL_V4.0.pkl', 'wb') as f:
    pickle.dump(pipe, f)

print("Model saved as RealVisXL_V4.0.pkl")
```

```python

from google.colab import drive
import os

drive.mount('/content/drive')

source_path = 'RealVisXL_V4.0.pkl'
destination_folder = '/content/drive/My Drive/Realviz_Model/'
destination_path = os.path.join(destination_folder, source_path)

os.makedirs(destination_folder, exist_ok=True)

import shutil
shutil.copyfile(source_path, destination_path)

print(f"'{source_path}' successfully saved to '{destination_path}'")
```

```python
##### SAFE TENSORS ##


# Fuse LoRA weights into the model
pipe.fuse_lora(lora_scale=1.0)

# Save the entire pipeline with safetensors serialization
pipe.save_pretrained("RealVisXL_V4.0_with_lora", safe_serialization=True)
```
